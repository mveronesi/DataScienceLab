```python
import csv

dataset = []
with open("../datasets/mnist.csv") as f:
    for row in csv.reader(f):
        dataset.append(row)
```

```python
def map_value_to_char(n):
    if(n>=0 and n<64):
        return " "
    else:
        if(n>=64 and n<128):
            return "."
        else:
            if(n>=128 and n<192):
                return "*"
            else:
                if(n>=192 and n<256):
                    return "#"
                else:
                    return "ERROR"
```

```python
k = -1
N = len(dataset)
while(k<=0 or k>N):
    k = int(input(f"inserisci k tra 1 e {N}: "))

row = dataset[k-1]
for i in range(0, 28):
    for j in range(0, 28):
        value = i*28 + j
        print(map_value_to_char(int(row[value])), end='')
    print()
```

```python
def euclidean_distance(k1, k2):
    v1 = dataset[k1-1]
    v2 = dataset[k2-1]
    if(len(v1) != len(v2)):
        print('Vectors have different dimensions!')
        return -1
    tot = 0
    for i in range(len(v1)):
        tot += (int(v1[i]) - int(v2[i]))**2
    tot **= 0.5
    return tot
```

```python
pos = [26, 30, 32, 35]
for i in range(0, len(pos)):
    for j in range(i+1, len(pos)):
        distance = euclidean_distance(pos[i],pos[j])
        print(f'Distance between {pos[i]} and {pos[j]}: {distance}')
```
