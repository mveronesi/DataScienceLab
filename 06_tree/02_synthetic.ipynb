```python
# 0) importing useful libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
import numpy as np
```

```python
# 1) dataset loading
df = pd.read_csv("../datasets/2d-synthetic.csv")
plt.figure(figsize=(15,15))
plt.scatter(df["x0"], df["x1"], c=df["label"], cmap="")
plt.show()
```

```python
# 2) model building
X = np.array([df.x0, df.x1]).T
y = np.array(df.label)
model = DecisionTreeClassifier()
model.fit(X, y)
plt.figure(figsize=(20,20))
tmp=plot_tree(model, feature_names=["x0", "x1"], fontsize=10)
```
